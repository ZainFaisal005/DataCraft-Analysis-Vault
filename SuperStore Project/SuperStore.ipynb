# **Super Store**

```python
import pandas as pd
```

```python
df = pd.read_csv('SampleSuperstore.csv')
df.head()
```

```python
df.columns
```

### Total amount of sales

```python
sales = df['Sales'].sum().round(2)
print('Total amount of sales is ',sales)
```

### Relationship between sales and profit

```python
print('Profit percentage in each sale')
pd.DataFrame((df['Profit'] / df['Sales']) * 100).head()
```

### Relationship between category and sales

```python
pd.DataFrame(df.groupby('Category')['Sales'].sum())
```

### Relationship between category and profit

```python
pd.DataFrame(df.groupby('Category')['Profit'].sum())
```

### Relationship between sales and city

```python
pd.DataFrame(df.groupby('City')['Sales'].sum().sort_values(ascending=False)).head()
```
